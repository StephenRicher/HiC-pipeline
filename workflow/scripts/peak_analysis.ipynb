```python
import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter, median_filter
```

```python
def readHomer(file, delimiter='\t'):
    homer = {}
    with open(file) as fh:
        homer['header'] = fh.readline()
        ncols = len(homer['header'].split(delimiter))
    homer['intervals'] = pd.read_csv(file, delimiter=delimiter, usecols=[0,1], header=0)
    homer['data'] = np.loadtxt(file, delimiter=delimiter, skiprows=1, usecols=range(2,ncols))
    return homer
```

```python
def writeHomer(homer):
    combined =  pd.concat([
        homer['intervals'], 
        pd.DataFrame(data=homer['data'])], axis=1)
    sys.stdout.write(homer['header'])
    combined.to_csv(sys.stdout, header=False, index=False, sep='\t')
```
